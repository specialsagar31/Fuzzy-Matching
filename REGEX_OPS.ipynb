```python
df["request_first_name"] = df["request_first_name"].astype(str)
df["response_name"] = df["response_name"].astype(str)

df["request_voter_id"] = df["request_voter_id"].astype(str)
df["response_voter"] = df["response_voter"].astype(str)

df["request_address_line1"] = df["request_address_line1"].astype(str)
df["response_address_line_1"] = df["response_address_line_1"].astype(str)

df["request_state"] = df["request_state"].astype(str)
df["response_state"] = df["response_state"].astype(str)

df["request_pin_code"] = df["request_pin_code"].astype(str)
df["response_pin_code"] = df["response_pin_code"].astype(str)

df["request_dob"] = df["request_dob"].astype(str)
df["response_dob"] = df["response_dob"].astype(str)
```

```python
from fuzzywuzzy import fuzz
l = []
for i,j in zip(df["request_first_name"],df["response_name"]):
    Str1 = i
    Str2 = j
    Ratio = fuzz.ratio(Str1.lower(),Str2.lower())
    l.append(Ratio)
df["score_name"] = l

m = []
for i,j in zip(df["request_voter_id"],df["response_voter"]):
    Str1 = i
    Str2 = j
    Ratio = fuzz.ratio(Str1.lower(),Str2.lower())
    m.append(Ratio)
df["score_voter_id"] = m

n = []
for i,j in zip(df["request_address_line1"],df["response_address_line_1"]):
    Str1 = i
    Str2 = j
    Ratio = fuzz.ratio(Str1.lower(),Str2.lower())
    n.append(Ratio)
df["score_address_line"] = n

o = []
for i,j in zip(df["request_state"],df["response_state"]):
    Str1 = i
    Str2 = j
    Ratio = fuzz.ratio(Str1.lower(),Str2.lower())
    o.append(Ratio)
df["score_state"] = o


p = []
for i,j in zip(df["request_pin_code"],df["response_pin_code"]):
    Str1 = i
    Str2 = j
    Ratio = fuzz.ratio(Str1.lower(),Str2.lower())
    p.append(Ratio)
df["score_pincode"] = p


q = []
for i,j in zip(df["request_dob"],df["response_dob"]):
    Str1 = i
    Str2 = j
    Ratio = fuzz.ratio(Str1.lower(),Str2.lower())
    q.append(Ratio)
df["score_dob"] = q
```
